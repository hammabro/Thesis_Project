# cat variables distribution : 

```python
# Visualising distribution of categorical variables 
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
sns.countplot(x='Carburant', data=df, ax=ax)
ax.set_xlabel('Carburant', fontsize=14)
ax.set_title('fuel type distribution', fontsize=14)
plt.xticks(fontsize=7 , rotation=20)  # Rotate x-axis labels by 45 degrees
# plt.savefig('count_carburant.png')
plt.show()
```

```python
# Visualising distribution of categorical variables 
fig, ax = plt.subplots(1,1, figsize=(8,5))
_ = sns.countplot(x='Boite Vitesse', data=df)
_ = plt.xlabel('Boite Vitesse', fontsize=14)
_ = plt.title('gearbox type distribution', fontsize=14)
plt.savefig('count_boitevit.png')
```

```python
# Visualising distribution of categorical variables 
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
sns.countplot(x='Carrosserie', data=df, ax=ax)
ax.set_xlabel('Carrosserie', fontsize=14)
ax.set_title('body type distribution', fontsize=14)
plt.xticks(rotation=20)  # Rotate x-axis labels by 45 degrees
plt.savefig('count_carrosserie.png')
plt.show()
```
